```python
from numpy import*
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

df=pd.read_csv("C:/Users/jinhcho/Downloads/Conductivtiy_experiment.csv",sep=";")
t=df['temperature']
c=df['EIS_conductivity']

t1=t[316:356]
t2=t[556:606]
t3=t[1745:1785]
t4=t[106:126]
t5=t[126:146]
t6=t[146:186]

c1=c[316:356]
c2=c[556:606]
c3=c[1745:1785]
c4=c[106:126]
c5=c[126:146]
c6=c[146:186]

def Cond(t,a,b):
    return a*t+b


par1, cov1 = curve_fit(Cond,t1,c1)
par2, cov2 = curve_fit(Cond,t2,c2)
par3, cov3 = curve_fit(Cond,t3,c3)
par4, cov4 = curve_fit(Cond,t4,c4)
par5, cov5 = curve_fit(Cond,t5,c5)
par6, cov6 = curve_fit(Cond,t6,c6)


plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.plot(t1, c1, 'b.',t1, Cond(t1,*par1), 'b')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['1.5:1.5:7'])

plt.subplot(3,2,5)
plt.plot(t2, c2, 'g.',t2, Cond(t2,*par2), 'g')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0:3:7'])

plt.subplot(3,2,3)
plt.plot(t3, c3, 'c.',t3, Cond(t3,*par3), 'c')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0.33:2.66:7'])

plt.subplot(3,2,2)
plt.plot(t4, c4, 'k.',t4, Cond(t4,*par4), 'k')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['2:3:5'])

plt.subplot(3,2,4)
plt.plot(t5, c5, 'r.',t5, Cond(t5,*par5), 'r')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['1:4:5'])

plt.subplot(3,2,6)
plt.plot(t6, c6, 'y.',t6, Cond(t6,*par6), 'y')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0:5:5'])

plt.tight_layout()
plt.show()

plt.figure()
plt.plot(t1, Cond(t1,*par1), 'b',label='1.5:1.5:7')
plt.plot(t3, Cond(t3,*par3), 'c',label='0.33:2.66:7')
plt.plot(t2, Cond(t2,*par2), 'g',label='0:3:7')
plt.plot(t4, Cond(t4,*par4), 'k',label='2:3:5')
plt.plot(t5, Cond(t5,*par5), 'r',label='1:4:5')
plt.plot(t6, Cond(t6,*par6), 'y',label='0:5:5')
plt.title("Conductivity")
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(loc=2)


plt.tight_layout()
plt.show()



plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.plot(t1, c1, 'b.',t1, Cond(t1,*par1), 'b')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['1.5:1.5:7'])

plt.subplot(2,3,3)
plt.plot(t2, c2, 'g.',t2, Cond(t2,*par2), 'g')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0:3:7'])

plt.subplot(2,3,2)
plt.plot(t3, c3, 'c.',t3, Cond(t3,*par3), 'c')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0.33:2.66:7'])

plt.subplot(2,3,4)
plt.plot(t4, c4, 'k.',t4, Cond(t4,*par4), 'k')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['2:3:5'])

plt.subplot(2,3,5)
plt.plot(t5, c5, 'r.',t5, Cond(t5,*par5), 'r')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['1:4:5'])

plt.subplot(2,3,6)
plt.plot(t6, c6, 'y.',t6, Cond(t6,*par6), 'y')
plt.ylabel("Conductivity (S/cm)")
plt.xlabel("Temperature (oC)")
plt.legend(['0:5:5'])

plt.tight_layout()
plt.show()
```
