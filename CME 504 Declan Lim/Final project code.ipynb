# 1st Fig

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#4.6% CO2
df = pd.read_csv('fig2a4.6.csv')
x1=df['x']
y1=df[' y']
x1data=np.array(x1)
y1data=np.array(y1)

#8.6% CO2
df = pd.read_csv('fig2a8.6.csv')
x2=df['x']
y2=df[' y']
x2data=np.array(x2)
y2data=np.array(y2)

#13.9% CO2
df = pd.read_csv('fig2a13.9.csv')
x3=df['x']
y3=df[' y']
x3data=np.array(x3)
y3data=np.array(y3)



xfit1=np.linspace(min(x1data),max(x1data),1000)
k1=interp1d(x1data,y1data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')

xfit2=np.linspace(min(x2data),max(x2data),1000)
k2=interp1d(x2data,y2data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')

xfit3=np.linspace(min(x3data),max(x3data),1000)
k3=interp1d(x3data,y3data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')


plt.figure(figsize=(5,5))
plt.xlim(0, 3)
plt.ylim(0, 550)
plt.plot(x1data,y1data,'bo')
plt.plot(xfit1,k1(xfit1), color='b', label='4.6% CO2')
plt.plot(x2data,y2data,'go')
plt.plot(xfit2,k2(xfit2), color='g', label='8.6% CO2')
plt.plot(x3data,y3data,'ro')
plt.plot(xfit3,k3(xfit3),color='r', label='13.9% CO2')
plt.xlabel("Cell voltage (V)")
plt.ylabel("Current density (mA cm-2)")
plt.title('The I–V curves of ORR/OER electrolysis under different CO2 concentrations')
plt.legend()
plt.show()



```

# 2nd Fig

```python
#Cubic
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

df = pd.read_csv('fig2bfec.csv')
x1=df['x']
y1=df[' y']
x1data=np.array(x1)
y1data=np.array(y1)

xfit=np.linspace(min(x1data),max(x1data),1000)
k=interp1d(x1data,y1data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')

plt.figure(figsize=(5,5))
plt.xlim(0, 550)
plt.ylim(0, 110)
plt.plot(x1data,y1data,'bo',xfit,k(xfit))
plt.xlabel("Current density (mA cm-2)")
plt.ylabel("FE_carbonate (%)")
plt.grid()
plt.show()
```

```python
#CubicSpline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

df = pd.read_csv('fig2bfec.csv')
x1=df['x']
y1=df[' y']
x1data=np.array(x1)
y1data=np.array(y1)

f=CubicSpline(x1data,y1data,bc_type='natural')

xplt=np.linspace(min(x1data),max(x1data),1000)
plt.plot(xplt,f(xplt),'b')
plt.plot(x1data,y1data,'bo', label = '13.9% CO2')

plt.xlabel("Current density (mA cm-2)")
plt.ylabel("FE_carbonate (%)")
plt.axhline(y=90, color='r', linestyle='-', label = "90%")
plt.xlim(0, 520)
plt.ylim(0, 110)
plt.title('The carbon-capture rate and FEcarbonate as a function of cell current density')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats

df = pd.read_csv('fig2eair.csv')
x1=df['x']
y1=df[' y']
x1data=np.array(x1)
y1data=np.array(y1)

df = pd.read_csv('fig2eo2.csv')
x2=df['x']
y2=df[' y']
x2data=np.array(x2)
y2data=np.array(y2)


xfit1=np.linspace(min(x1data),max(x1data),1000)
k1=interp1d(x1data,y1data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')

xfit2=np.linspace(min(x2data),max(x2data),1000)
k2=interp1d(x2data,y2data,kind = 'cubic',bounds_error = False, fill_value = 'extrapolate')

plt.figure(figsize=(5,5))
plt.xlim(1.2, 3.2)
plt.ylim(0, 550)
plt.plot(x1data,y1data,'bo')
plt.plot(xfit1,k1(xfit1), color='b', label='Air + 13.9% СО2')
plt.plot(x2data,y2data,'ro')
plt.plot(xfit2,k2(xfit2), color='r', label='O2 + 13.9% СО2')
plt.xlabel("Cell voltage (V)")
plt.ylabel("Current density (mA cm-2)")
plt.legend()
plt.title('Carbon-capture performance comparisons between O2 and air carrier gas')
plt.show()
```
