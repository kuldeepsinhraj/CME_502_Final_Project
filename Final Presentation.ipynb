```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data
data = {'Puddle Sample Location': ['1 Control Funnel', '2 Fort Tunnel', '3 C&D Parking', '4 Track & Field', 
                                   '5 Grassy Field', '6 Pennyfield Ave'],
    'Enterococcus MPN/100mL Day 1': [30, 420, 1090, 384, 4106, 1370],
    'Enterococcus MPN/100mL Day 2': [31, 738, 990, 41, 496, 2415]}

# Create DataFrame from data
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Set width of bars
bar_width = 0.35

# Set locations for bars
index = np.arange(len(df['Puddle Sample Location']))

# Plotting bars for Day 1
plt.bar(index - bar_width/2, df['Enterococcus MPN/100mL Day 1'], bar_width, label='Day 1', color='brown')

# Plotting bars for Day 2
plt.bar(index + bar_width/2, df['Enterococcus MPN/100mL Day 2'], bar_width, label='Day 2', color='green')

# Add labels, title, legend, and grid
plt.xlabel('Puddle Sample Location')
plt.ylabel('Enterococcus MPN/100mL')
plt.title('Enterococcus MPN Values in Puddle Samples')
plt.xticks(index, df['Puddle Sample Location'])
plt.legend()
plt.grid(axis='y')

# Show plot
plt.tight_layout()
plt.show()
```
