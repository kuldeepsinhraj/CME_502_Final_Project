```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = {
    '% NaOH': [40, 50, 40, 50, 37.9, 52.1, 45, 45, 45, 45, 45],
    'Weight of MCA(g)': [12, 12, 18, 18, 15, 15, 10.7, 19.2, 15, 15, 15],
    '% Yield': [160.0, 163.4, 141.8, 170.2, 140.0, 159.8, 150.4, 156.0, 162.6, 161.6, 160.4],
    'DS': [0.76, 0.92, 0.74, 0.95, 0.57, 0.69, 0.71, 0.69, 0.86, 0.82, 0.83]
}

# Plot area sizing
plt.figure(figsize=(18, 5))

# % NaOH vs % Yield
plt.subplot(1, 3, 1)
plt.scatter(data['% NaOH'], data['% Yield'], color='blue')
plt.title('% NaOH vs % Yield')
plt.xlabel('% NaOH')
plt.ylabel('% Yield')

# % NaOH vs DS
plt.subplot(1, 3, 2)
plt.scatter(data['% NaOH'], data['DS'], color='red')
plt.title('% NaOH vs DS')
plt.xlabel('% NaOH')
plt.ylabel('DS')

# Weight of MCA(g) vs % Yield
plt.subplot(1, 3, 3)
plt.scatter(data['Weight of MCA(g)'], data['% Yield'], color='green')
plt.title('Weight of MCA(g) vs % Yield')
plt.xlabel('Weight of MCA(g)')
plt.ylabel('% Yield')

plt.tight_layout()
plt.show()

# Weight of MCA(g) vs DS
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.scatter(data['Weight of MCA(g)'], data['DS'], color='orange')
plt.title('Weight of MCA(g) vs DS')
plt.xlabel('Weight of MCA(g)')
plt.ylabel('DS')

# Contour plot for % Yield
plt.subplot(1, 3, 2)
plt.tricontourf(data['% NaOH'], data['Weight of MCA(g)'], data['% Yield'], cmap='viridis')
plt.colorbar(label='% Yield')
plt.scatter(data['% NaOH'], data['Weight of MCA(g)'], color='blue', marker='o')
plt.title('% Yield Contour Plot')
plt.xlabel('% NaOH')
plt.ylabel('Weight of MCA(g)')

# Contour plot for DS
plt.subplot(1, 3, 3)
plt.tricontourf(data['% NaOH'], data['Weight of MCA(g)'], data['DS'], cmap='plasma')
plt.colorbar(label='DS')
plt.scatter(data['% NaOH'], data['Weight of MCA(g)'], color='red', marker='o')
plt.title('DS Contour Plot')
plt.xlabel('% NaOH')
plt.ylabel('Weight of MCA(g)')

plt.tight_layout()
plt.show()

# Find highest % Yield and DS
max_yield_index = data['% Yield'].index(max(data['% Yield']))
max_ds_index = data['DS'].index(max(data['DS']))

print(f'Highest % Yield: {max(data["% Yield"])} at % NaOH = {data["% NaOH"][max_yield_index]} and Weight of MCA(g) = {data["Weight of MCA(g)"][max_yield_index]}')
print(f'Highest DS: {max(data["DS"])} at % NaOH = {data["% NaOH"][max_ds_index]} and Weight of MCA(g) = {data["Weight of MCA(g)"][max_ds_index]}')
```
